```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import os
```

# Generar Data Simulada
#### Usaremos el mismo Proceso Generador de Datos (PGD) que en el LAB2

$
\begin{equation}
    Y = \exp{(4X)} + \varepsilon, \quad \varepsilon \sim N(0,1)
\end{equation}
$


```python
### Generar Data simulada X e Y
X= np.random.uniform(0,1,1000)
X.sort()
X= X.reshape(-1,1)

e= np.random.uniform(0,1,1000)
e=e.reshape(-1,1)

Y= np.exp(4 * X) + e
```

# Generar modelo con X1=X


```python

#R cuadrado y R ajustado

model = LinearRegression(fit_intercept=False)
model.fit(X,Y)
Y_hat= model.predict(X)

#R squared

mse_test = np.mean( (Y-Y_hat)**2)
R_sq = 1 - mse_test / np.mean((Y-np.mean(Y))** 2)


#Adjusted R squared
adj_mse_test = 1000 / (1000-1) * mse_test
adjR_sq_test = 1 - adj_mse_test / np.mean((Y-np.mean(Y))** 2)


#R-squared-Out of Sample
X1=X
X_train, X_test,Y_train, Y_test = train_test_split(X1,Y) 

model = LinearRegression(fit_intercept=False)
model.fit(X_train,Y_train)
Y_hat_test = model.predict(X_test)
mse_1_test = np.mean( (Y_test-Y_hat_test)**2)
R_sq_outofsample = 1 - mse_1_test / np.mean((Y_test-np.mean(Y_test))** 2)

print(f'R cuadrado es {R_sq}')
print(f'R ajustado es {adjR_sq_test}')
print(f'R out of sample es {R_sq_outofsample}')

```

# Generar modelo con dos features X1=X , x2=X^2


```python
X_2 = np.hstack((X,X**2))

model = LinearRegression(fit_intercept=False)
model.fit(X_2,Y)
Y_hat= model.predict(X_2)

#R squared

mse_test = np.mean( (Y-Y_hat)**2)
R_sq = 1 - mse_test / np.mean((Y-np.mean(Y))** 2)


#Adjusted R squared
adj_mse_test = 1000 / (1000-1) * mse_test
adjR_sq_test = 1 - adj_mse_test / np.mean((Y-np.mean(Y))** 2)


#R-squared-Out of Sample
X_2_train, X_2_test,Y_2_train, Y_2_test = train_test_split(X_2,Y) 

model2 = LinearRegression(fit_intercept=False)
model2.fit(X_2_train,Y_2_train)
Y_2_hat_test = model2.predict(X_2_test)

mse_2_test = np.mean( (Y_2_test-Y_2_hat_test)**2)
R_sq_outofsample = 1 - mse_2_test / np.mean((Y_2_test-np.mean(Y_2_test))** 2)


print(f'R cuadrado es {R_sq}')
print(f'R ajustado es {adjR_sq_test}')
print(f'R out of sample es {R_sq_outofsample}')
```

# Generar un Loop con features crecientes (1,2,5,10...)

```python
Features= [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
### En este caso
Features[0]
R_sq_Lista = []
adjustR_sq_Lista = []
R_squared_OutofSample_List =[]
# Calcular R cuadrado y R ajustado
#Definimos X

for k in Features:
    # construye una lista de arrays: X^1, X^2, ..., X^k
    X_list = [X**i for i in range(1, k+1)]
    X_i = np.hstack((X_list))
    # generar modelo
    model = LinearRegression(fit_intercept=False)
    model.fit(X_i,Y)
    Y_hat= model.predict(X_i)

    #R CUADRADO
    mse_test = np.mean( (Y-Y_hat)**2)
    R_sq = 1 - mse_test / np.mean((Y-np.mean(Y))** 2)
    R_sq_Lista.append(R_sq)

    #R CUADRADO AJUSTAD
    if k != 1000:
        adj_mse_test = 1000 / (1000-k) * mse_test
        adjR_sq_test = 1 - adj_mse_test / np.mean((Y-np.mean(Y))** 2)
        
    else:
        adjR_sq_test = 1
    adjustR_sq_Lista.append(adjR_sq_test)
    #R-squared-Out of Sample
    X_train, X_test,Y_train, Y_test = train_test_split(X_i ,Y) 

    model = LinearRegression(fit_intercept=False)
    model.fit(X_train,Y_train)
    Y_hat_test = model.predict(X_test)
    mse_1_test = np.mean( (Y_test-Y_hat_test)**2)
    R_sq_outofsample = 1 - mse_1_test / np.mean((Y_test-np.mean(Y_test))** 2)
    R_squared_OutofSample_List.append(R_sq_outofsample )


```

## Observación Importante 
### Teóricamente dado que es un modelo sin intercepto debería de usarse el R-Cuadrado Descentrado. En casos de modelos que no tienen intercepto el no usarlo puede llevar a errores como valores de R cuadrado negativos o incluso mayores que 1, por lo que dado esto usaremos el R cuadrado descentrado, para los gráficos. Además en los demás lenguajes de programación se usará directamente el R cuadrado descentrado.

```python
Features= [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
### En este caso
Features[0]
R_sq_Lista = []
adjustR_sq_Lista = []
R_squared_OutofSample_List =[]
# Calcular R cuadrado y R ajustado
#Definimos X

for k in Features:
    # construye una lista de arrays: X^1, X^2, ..., X^k
    X_list = [X**i for i in range(1, k+1)]
    X_i = np.hstack((X_list))
    # generar modelo
    model = LinearRegression(fit_intercept=False)
    model.fit(X_i,Y)
    Y_hat= model.predict(X_i)

    #R CUADRADO
    mse_test = np.mean( (Y-Y_hat)**2)
    R_sq = 1 - mse_test / np.mean((Y)** 2)
    R_sq_Lista.append(R_sq)

    #R CUADRADO AJUSTAD
    if k != 1000:
        adj_mse_test = 1000 / (1000-k) * mse_test
        adjR_sq_test = 1 - adj_mse_test / np.mean((Y)** 2)
        
    else:
        adjR_sq_test = 1
    adjustR_sq_Lista.append(adjR_sq_test)
    #R-squared-Out of Sample
    X_train, X_test,Y_train, Y_test = train_test_split(X_i ,Y) 

    model = LinearRegression(fit_intercept=False)
    model.fit(X_train,Y_train)
    Y_hat_test = model.predict(X_test)
    mse_1_test = np.mean( (Y_test-Y_hat_test)**2)
    R_sq_outofsample = 1 - mse_1_test / np.mean((Y_test)** 2)
    R_squared_OutofSample_List.append(R_sq_outofsample )
```
